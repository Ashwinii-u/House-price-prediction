#### Step 4. Removing Otliers

```python
house_df = df.drop(['id','date'], axis = 1)
```

```python
from scipy.stats import zscore
z = np.abs(zscore(house_df))
```

```python
len(np.where((z > 3))[0]) / house_df.shape[0]*100
```

```python
len(np.where((z > 4))[0]) / house_df.shape[0]*100
```

```python
Q1 = house_df['price'].quantile(0.25)
Q3 = house_df['price'].quantile(0.75)
# Inter-Quartile Range
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['price'] = df['price'].clip(lower =lower_bound, upper = upper_bound)
```

```python
sns.boxplot(df['price'], color = 'skyblue', orient = 'h')
```

```python
sns.distplot(house_df['price'], color = 'blue', fit = norm)
plt.show()
```

#### Step 5. is there any multi-collinearity ?

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
x = house_df.drop(['price'], axis = 1)
x = x.astype(float)
# Empty Dataframe
vif_data = pd.DataFrame()
vif_data['Features'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
```

```python
vif_data
```

##### Step 6. Features and Label

```python
X = house_df.drop(['sqft_above','sqft_living','sqft_basement','long','zipcode','yr_renovated','price'], axis = 1)
Y = house_df['price']
```

#### Step 7. Cross Validation
    * Split Data into Train & Test Split

```python
from sklearn.model_selection import train_test_split
# test_size = 0.2, This indicates train_sample = 80% and test_sample_size = 20%
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

#### Step 8. Linear Regression

```python
from sklearn.linear_model import LinearRegression
# initialize model
lr_model = LinearRegression()
```

```python
# fit linear regression model
lr_model.fit(x_train, y_train)
```

```python
# R-Squared Score (0-1)
lr_model.score(x_test, y_test)
```

```python
lr_model.coef_
```

```python
# beta0
lr_model.intercept_
```

#### Step 9. Regression Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_pred = lr_model.predict(x_test)
```

```python
# mean squared error (MSE)
mean_squared_error(y_test, y_pred)
```

```python
# Root mean square error (RMSE)
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r2_score(y_test,y_pred)
```
